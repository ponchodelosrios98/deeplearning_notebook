# Model Validation

To know how many epochs are just right for a training dataset, the dataset should have 3 sets:
    
    - Training Set (Update Model weights)
    - Validation Set (Check how well the model generalizes)
    - Test Set (Check accuracy of the model after it trains)
    


The validation set yells us when to stop training a model; when the validation loss stops decreasing (and especially when the validation loss starts increasing and the training loss is still decreasing)
